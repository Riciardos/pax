```python
import numpy as np
import pandas
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def slurp_df_responsibly(store, df_name, chunksize=10000, stop=float('inf')):
    """Slurp a dataframe from store into a numpy record array
    """
    # Dirty hack to get number of rows out so progress bar works
    n_rows = getattr(store._handle.root, df_name).table.nrows
    records = []
    for df in tqdm(store.select(df_name, 
                                stop=stop, 
                                chunksize=chunksize), 
                   total=int(min(stop,n_rows)/chunksize)):
        temp_records = df.to_records()
        records.append(temp_records)
    return np.concatenate(records)
```

```python
store = pandas.HDFStore('../../bin/output.hdf')
```

```python
recposes = slurp_df_responsibly(store, 'ReconstructedPosition')
```

```python
ws = recposes[recposes['algorithm'] == 'PosRecWeightedSum']
nn = recposes[recposes['algorithm'] == 'X100NeuralNet']
```

```python
#plt.figure(figsize=(12,10))
for x, y, title in (
    (ws['x'], ws['y'], 'Weighted sum'),
    (nn['x'], nn['y'], 'Neural net'),
    (ws['x'] - nn['x'], ws['y'] - nn['y'], 'Weighted sum - Neural net'),
):
    plt.figure(figsize=(7,7))
    plt.scatter(x, y, edgecolors='none', s=10, alpha=0.5,)
    plt.title(title)
    plt.show()
```
